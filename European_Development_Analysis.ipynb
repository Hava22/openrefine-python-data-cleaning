# European Development Indicators Analysis
## UN SDG Framework - Portfolio Project

**Objective**: Analyze development indicators across European countries using simulated SDG-like data  
**Skills Demonstrated**: Data analysis, visualization, statistical testing, portfolio development

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Setup for better visuals
plt.style.use('default')
sns.set_palette("husl")

print("‚úÖ Libraries imported successfully!")
```

```python
# Generate realistic European development data
np.random.seed(42)  # For reproducible results

countries = ['Germany', 'France', 'Italy', 'Spain', 'Poland', 'Netherlands', 'Sweden', 'Albania']
years = list(range(2015, 2023))

data = []
for country in countries:
    for year in years:
        data.append({
            'country': country,
            'year': year,
            'gdp_per_capita': max(10000, np.random.normal(30000, 10000)),
            'education_budget_pct': np.random.uniform(3, 7),
            'renewable_energy_pct': np.random.uniform(10, 50),
            'poverty_rate': np.random.uniform(5, 25)
        })

df = pd.DataFrame(data)
print(f"‚úÖ Dataset created: {len(df)} records")
print(f"üåç Countries: {df['country'].nunique()}")
print(f"üìÖ Years: {df['year'].min()}-{df['year'].max()}")
df.head()
```

## Data Exploration and Visualization

```python
# GDP trends visualization
plt.figure(figsize=(12, 6))
for country in df['country'].unique():
    country_data = df[df['country'] == country]
    plt.plot(country_data['year'], country_data['gdp_per_capita'], label=country, marker='o')

plt.title('GDP per Capita Trends in European Countries (2015-2022)')
plt.xlabel('Year')
plt.ylabel('GDP per Capita (USD)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Renewable energy comparison (2022 data)
latest_data = df[df['year'] == 2022]

plt.figure(figsize=(12, 6))
sns.barplot(data=latest_data, x='country', y='renewable_energy_pct')
plt.title('Renewable Energy Adoption by Country (2022)')
plt.xlabel('Country')
plt.ylabel('Renewable Energy (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
numeric_cols = ['gdp_per_capita', 'education_budget_pct', 'renewable_energy_pct', 'poverty_rate']

plt.figure(figsize=(10, 8))
correlation_matrix = df[numeric_cols].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, fmt='.2f', cbar_kws={'shrink': .8})
plt.title('Correlation Between Development Indicators')
plt.tight_layout()
plt.show()

print("üîó Key correlations:")
print("- GDP vs Poverty rate: Strong negative correlation")
print("- Education vs Renewable energy: Positive correlation")
```

## SDG Progress Assessment

```python
# SDG Progress Analysis
print("üéØ SDG PROGRESS ASSESSMENT (2015-2022)")
print("=" * 50)

# Map indicators to SDGs
sdg_mapping = {
    'gdp_per_capita': 'SDG 8: Economic Growth',
    'education_budget_pct': 'SDG 4: Quality Education', 
    'renewable_energy_pct': 'SDG 7: Clean Energy',
    'poverty_rate': 'SDG 1: No Poverty'
}

# Calculate progress
progress_data = []
for indicator, sdg in sdg_mapping.items():
    baseline_2015 = df[df['year'] == 2015][indicator].mean()
    current_2022 = df[df['year'] == 2022][indicator].mean()
    
    # For poverty rate, negative progress is good
    if indicator == 'poverty_rate':
        progress_pct = ((baseline_2015 - current_2022) / baseline_2015) * 100
    else:
        progress_pct = ((current_2022 - baseline_2015) / baseline_2015) * 100
    
    progress_data.append({
        'SDG': sdg,
        'Progress_Pct': round(progress_pct, 1)
    })

progress_df = pd.DataFrame(progress_data)
progress_df
```

```python
# SDG Progress Visualization
plt.figure(figsize=(12, 6))

colors = ['green' if x > 0 else 'red' for x in progress_df['Progress_Pct']]
bars = plt.barh(progress_df['SDG'], progress_df['Progress_Pct'], color=colors, alpha=0.7)

plt.axvline(x=0, color='black', linestyle='-', alpha=0.3)
plt.xlabel('Progress Since 2015 (%)')
plt.title('SDG Progress Assessment (2015-2022)')
plt.grid(True, alpha=0.3, axis='x')

# Add value labels
for bar, progress in zip(bars, progress_df['Progress_Pct']):
    plt.text(bar.get_width() + (1 if progress > 0 else -5), 
             bar.get_y() + bar.get_height()/2, 
             f'{progress}%', 
             ha='left' if progress > 0 else 'right', 
             va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```
